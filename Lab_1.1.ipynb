```python
img1 = img[:, :, 1]
```

```python
plt.imshow(img1, cmap='gray')
```

```python
h = np.zeros(256, np.int32)
for i in range(img1.shape[0]):
    for j in range(img1.shape[1]):
        for k in range(256):
            if img1[i][j] == k:
                h[k] += 1
for i in range(247):        
    print(i, h[i])

# print(arr)
```
