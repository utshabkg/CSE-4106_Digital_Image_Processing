```python
import numpy as np
from scipy import ndimage
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('grill.jpg')
```

```python
image = image[:,:,0]
plt.imshow(image)
print(image.shape)
```

```python
dx = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])
dy = np.array([[-1,-1,-1],[0,0,0],[1,1,1]])
vertical_x = ndimage.convolve(image, dx)
vertical_y = ndimage.convolve(image, dy)
vertical_xy = ndimage.convolve(vertical_x, dy)
vertical_yx = ndimage.convolve(vertical_y, dx)
```

```python
plt.imshow(vertical_x)
```

```python
plt.imshow(vertical_y)
```

```python
plt.imshow(vertical_xy)
```

```python
plt.imshow(vertical_yx)
```
