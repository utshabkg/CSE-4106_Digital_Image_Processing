##Imports


```python
import cv2
import matplotlib.pyplot as plt
import numpy as np


path = 'chess.jpg'
image = cv2.imread(path, 0)
plt.imshow(image)
```

```python
image.shape
```

```python
def imfilter(image, kernel):
    kernel = np.flipud(np.fliplr(kernel))
    output = np.zeros_like(image)

    image_padded = np.zeros((image.shape[0] + 2, image.shape[1] + 2))
    image_padded[1:-1, 1:-1] = image

  
    for x in range(image.shape[1]):
        for y in range(image.shape[0]):
            output[y, x] = (kernel * image_padded[y: y+3, x: x+3]).sum()

    return output
```

## Prewitt Operator

```python
vertical_Prewitt = np.array([
               [-1, 0, 1],
               [-1, 0, 1],
               [-1, 0, 1] ])

horizontal_Prewitt = np.array([
               [-1, -1, -1],
               [0, 0, 0],
               [1, 1, 1] ])
```

```python
Diff_inx = imfilter(image, horizontal_Prewitt)
Diff_iny = imfilter(image, vertical_Prewitt)

Diff_inxy = imfilter(Diff_inx, vertical_Prewitt)
Diff_inyx = imfilter(Diff_iny, horizontal_Prewitt)
```

```python
print("Horizontal Edge")
plt.imshow(Diff_inx)
```

```python
print("Vertical Edge")
plt.imshow(Diff_iny)
```

```python
print("Combine XY")
plt.imshow(Diff_inxy)
```

```python
print("Combine YX")
plt.imshow(Diff_inyx)
```

## Sobel Operator

```python
vertical_Sobel = np.array([
               [-1, 0, 1],
               [-2, 0, 2],
               [-1, 0, 1] ])

horizontal_Sobel = np.array([
               [-1, -2, -1],
               [0, 0, 0],
               [1, 2, 1] ])
```

```python
Diff_inx = imfilter(image, horizontal_Sobel)
Diff_iny = imfilter(image, vertical_Sobel)

Diff_inxy = imfilter(Diff_inx, vertical_Sobel)
Diff_inyx = imfilter(Diff_iny, horizontal_Sobel)
```

```python
print("Horizontal Edge")
plt.imshow(Diff_inx)
```

```python
print("Vertical Edge")
plt.imshow(Diff_iny)
```

```python
print("Combine XY")
plt.imshow(Diff_inxy)
```

```python
print("Combine YX")
plt.imshow(Diff_inyx)
```

## Laplacian Operator

```python
laplacian = np.array([
               [1, 1, 1],
               [1, -8, 1],
               [1, 1, 1] ])
```

```python
Laplacian = imfilter(image, laplacian)

print("Laclacian")
plt.imshow(Laplacian)
```
